# Section 1: Descriptive Statistics & Initial Exploration

##Part II

```python
#calculate the mean, median, mode of alcohol
alcohol_mean = winequality_df['alcohol'].mean()
alcohol_median = winequality_df['alcohol'].median()
alcohol_mode = winequality_df['alcohol'].mode()[0]

#calculate the mean, median, mode of residual sugar
residual_sugar_mean = winequality_df['residual sugar'].mean()
residual_sugar_median = winequality_df['residual sugar'].median()
residual_sugar_mode = winequality_df['residual sugar'].mode()[0]

#calculate the mean, median, mode of quality
quality_mean = winequality_df['quality'].mean()
quality_median = winequality_df['quality'].median()
quality_mode = winequality_df['quality'].mode()[0]

#print the result
print(f"""Alcohol Statistics
Alcohol mean : {alcohol_mean:.2f}
Alcohol median : {alcohol_median}
Alcohol mode : {alcohol_mode}

Residual Sugar Statistics
Residual Sugar mean : {residual_sugar_mean:.2f}
Residual Sugar median : {residual_sugar_median}
Residual Sugar mode : {residual_sugar_mode}

Quality Statistics
Quality mean : {quality_mean:.2f}
Quality median : {quality_median}
Quality mode : {quality_mode}
""")
```

```python
#calculate variance and standard deviation of alcohol
alcohol_var = winequality_df['alcohol'].var()
alcohol_std = winequality_df['alcohol'].std()

#calculate variance and standard deviation of residual sugar
residual_sugar_var = winequality_df['residual sugar'].var()
residual_sugar_std = winequality_df['residual sugar'].std()

#calculate variance and standard deviation of quality
quality_var = winequality_df['quality'].var()
quality_std = winequality_df['quality'].std()

#print the result
print(f"""Alcohol Statistics
Alcohol variance : {alcohol_var:.2f}
Alcohol standard deviation : {alcohol_std:.2f}

Residual Sugar Statistics
Residual Sugar variance : {residual_sugar_var:.2f}
Residual Sugar standard deviation : {residual_sugar_std:.2f}

Quality Statistics
Quality variance : {quality_var:.2f}
Quality standard deviation : {quality_std:.2f}
""")
```

##Part III

```python
#calculate total data and percentage of alcohol data
#calculate the mean first
alcohol_mean = winequality_df['alcohol'].mean()

#calculate the amount of data with alcohol > mean
alcohol_above = (winequality_df['alcohol'] > alcohol_mean).sum()

#calculate the length of wine quality data
total = len(winequality_df)

#make the percentage
alcohol_percentage = (alcohol_above/total)*100

#print the result
print(f"""Alcohol mean : {alcohol_mean:.2f}
Amount of alcohol data above the mean : {alcohol_above}
Percentage of alcohol data above the mean : {alcohol_percentage:.2f}%
""")
```

```python
#calculate total data and percentage of residual sugar data
#calculate the mean first
residual_sugar_mean = winequality_df['residual sugar'].mean()

#calculate the amount of data with residual sugar > mean
residual_sugar_above = (winequality_df['residual sugar'] > residual_sugar_mean).sum()

#calculate the length of wine quality data
total = len(winequality_df)

#make the percentage
residual_sugar_percentage = (residual_sugar_above/total)*100

#print the result
print(f"""Residual Sugar mean : {residual_sugar_mean:.2f}
Amount of residual sugar data above the mean : {residual_sugar_above}
Percentage of residual sugar data above the mean : {residual_sugar_percentage:.2f}%
""")
```

```python
#calculate total data and percentage of quality data
#calculate the mean first
quality_mean = winequality_df['quality'].mean()

#calculate the amount of data with quality > mean
quality_above = (winequality_df['quality'] > quality_mean).sum()

#calculate the length of wine quality data
total = len(winequality_df)

#make the percentage
quality_percentage = (quality_above/total)*100

#print the result
print(f"""Quality mean : {quality_mean:.2f}
Amount of quality data above the mean : {quality_above}
Percentage of quality data above the mean : {quality_percentage:.2f}%
""")
```

##Part IV

```python
print("Analytical Thinking Answers")
print()

#1. Which column has the most scattered values?
std = {
    "Alcohol" : alcohol_std,
    "Residual Sugar" : residual_sugar_std,
    "Quality" : quality_std
}

max_std = max(std, key=std.get)
print("1. The column with the most scattered values is", max_std)
print()


#2. Is there more data above or below the average?
print("2.")
percentages = {
    "Alcohol" : alcohol_percentage,
    "Residual Sugar" : residual_sugar_percentage,
    "Quality" : quality_percentage
}

for col, pct in percentages.items():
  if pct >= 50:
    pct_status = "above the mean"
  else:
    pct_status = "below the mean"
  print(f"The data from {col} is mostly {pct_status}. ({pct:.2f}%)")
print()


#3. Is the data symmetrical or skewed?
print("3.")
data = {
    "Alcohol" : [alcohol_mean, alcohol_median, alcohol_std],
    "Residual Sugar" : [residual_sugar_mean, residual_sugar_median, residual_sugar_std],
    "Quality" : [quality_mean, quality_median, quality_std]
}

for col, metrics in data.items():
  mean, median, std = metrics
  diff = abs(mean - median)
  if diff < 0.1:
    skewed = "roughly symmetric"
  elif mean > median:
    skewed ="right skewed"
  else:
    skewed = "left skewed"
  print(f"The distribution for {col} column is {skewed} because mean = {mean:.2f}, median = {median:.2f}, difference = {diff:.2f}")
print()

#4.Is the data stable or does it vary widely?
print("4.")

for col, metrics in data.items():
  mean, median, std = metrics
  if std < mean * 0.2:
    distribution = "fairly consistent"
  else:
    distribution = "more scattered"
  print(f"The data from {col} is {distribution} because mean = {mean:.2f}, std = {std:.2f}")

print()
print("Analytical Thinking Summary/Insight")
print()
print("""Based on the analysis above, the Residual Sugar column has the most varied values compared to Alcohol and Quality,
which means its distribution is wider and less consistent. While, Alcohol and Quality are fairly consistent.
When viewed from the average, Alcohol and Residual Sugar tend to fall below the mean, while Quality is slightly above the mean.
In terms of distribution, Alcohol and Residual Sugar are right skewed (mean greater than median),
whereas Quality is left skewed (mean less than median).
""")
```

#Session 2: Analysis of the Relationship between Variables

##Part I

```python
#calculate the covariance and correlation between alcohol & quality
cov_value_1 = winequality_df['alcohol'].cov(winequality_df['quality'])
corr_value_1 = winequality_df['alcohol'].corr(winequality_df['quality'])

#calculate the covariance and correlation between residual sugar & quality
cov_value_2 = winequality_df['residual sugar'].cov(winequality_df['quality'])
corr_value_2 = winequality_df['residual sugar'].corr(winequality_df['quality'])

#calculate the covariance and correlation between density & residual sugar
cov_value_3 = winequality_df['density'].cov(winequality_df['residual sugar'])
corr_value_3 = winequality_df['density'].corr(winequality_df['residual sugar'])

#print the result
print(f"""The covariance between alcohol & quality is {cov_value_1:.2f}
The correlation between alcohol & quality is {corr_value_1:.2f}

The covariance between residual sugar & quality is {cov_value_2:.2f}
The correlation between residual sugar & quality is {corr_value_2:.2f}

The covariance between density & residual sugar is {cov_value_3:.2f}
The correlation between density & residual sugar is {corr_value_3:.2f}
""")
```

##Part II

```python
#manual calculation of covariance and correlation between alcohol & quality

#take the first row of data
x1_1 = winequality_df['alcohol'].iloc[0]
y1_1 = winequality_df['quality'].iloc[0]

#calculate the mean
x_mean_1 = winequality_df['alcohol'].mean()
y_mean_1 = winequality_df['quality'].mean()

#calculate the difference
x_diff_1 = x1_1 - x_mean_1
y_diff_1 = y1_1 - y_mean_1

#calculate the covariance
cov_1 = x_diff_1 * y_diff_1

#take the standard deviation
x_std_1 = winequality_df['alcohol'].std()
y_std_1 = winequality_df['quality'].std()

#calculate the correlation
corr_1 = (x_diff_1 / x_std_1) * (y_diff_1 / y_std_1)

#print the result
print(f"""Manual Calculation of Covariance and Correlation between Alcohol & Quality
Sample data (x1, y1) = ({x1_1}, {y1_1})
Mean alcohol = {x_mean_1:.2f}
Mean quality = {y_mean_1:.2f}
(x1 - mean_x) = {x_diff_1:.2f}
(y1 - mean_y) = {y_diff_1:.2f}
Std alcohol = {x_std_1:.2f}
Std quality = {y_std_1:.2f}
Covariance = {cov_1:.2f}
Correlation = {corr_1:.2f}
""")
```

```python
#manual calculation of covariance and correlation between residual sugar & quality

#take the first row of data
x1_2 = winequality_df['residual sugar'].iloc[0]
y1_2 = winequality_df['quality'].iloc[0]

#calculate the mean
x_mean_2 = winequality_df['residual sugar'].mean()
y_mean_2 = winequality_df['quality'].mean()

#calculate the difference
x_diff_2 = x1_2 - x_mean_2
y_diff_2 = y1_2 - y_mean_2

#calculate the covariance
cov_2 = x_diff_2 * y_diff_2

#take the standard deviation
x_std_2 = winequality_df['residual sugar'].std()
y_std_2 = winequality_df['quality'].std()

#calculate the correlation
corr_2 = (x_diff_2 / x_std_2) * (y_diff_2 / y_std_2)

#print the result
print(f"""Manual Calculation of Covariance and Correlation between Residual Sugar & Quality
Sample data (x1, y1) = ({x1_2}, {y1_2})
Mean residual sugar = {x_mean_2:.2f}
Mean quality = {y_mean_2:.2f}
(x1 - mean_x) = {x_diff_2:.2f}
(y1 - mean_y) = {y_diff_2:.2f}
Std alcohol = {x_std_2:.2f}
Std quality = {y_std_2:.2f}
Covariance = {cov_2:.2f}
Correlation = {corr_2:.2f}
""")
```

```python
#manual calculation of covariance and correlation between density & residual sugar

#take the first row of data
x1_3 = winequality_df['density'].iloc[0]
y1_3 = winequality_df['residual sugar'].iloc[0]

#calculate the mean
x_mean_3 = winequality_df['density'].mean()
y_mean_3 = winequality_df['residual sugar'].mean()

#calculate the difference
x_diff_3 = x1_3 - x_mean_3
y_diff_3 = y1_3 - y_mean_3

#calculate the covariance
cov_3 = x_diff_3 * y_diff_3

#take the standard deviation
x_std_3 = winequality_df['density'].std()
y_std_3 = winequality_df['residual sugar'].std()

#calculate the correlation
corr_3 = (x_diff_3 / x_std_3) * (y_diff_3 / y_std_3)

#print the result
print(f"""Manual Calculation of Covariance and Correlation between Density & Residual Sugar
Sample data (x1, y1) = ({x1_3}, {y1_3})
Mean density = {x_mean_3:.2f}
Mean residual sugar = {y_mean_3:.2f}
(x1 - mean_x) = {x_diff_3:.2f}
(y1 - mean_y) = {y_diff_3:.2f}
Std alcohol = {x_std_3:.2f}
Std quality = {y_std_3:.2f}
Covariance = {cov_3:.2f}
Correlation = {corr_3:.2f}
""")
```

#Sessison 3: Z-Score Analysis

##Part I

```python
#calculate the mean & std
quality_mean = winequality_df['quality'].mean()
quality_std = winequality_df['quality'].std()

#calculate the z score
winequality_df['quality_z_score'] = (winequality_df['quality'] - quality_mean) / quality_std

#print the result
print(f"Quality Mean : {quality_mean:.2f}")
print(f"Quality Standard Deviation : {quality_std:.2f}")
print("Quality Z-Score :")
print(winequality_df['quality_z_score'].head(10))
```

##Part II

```python
#1. How many wines have a Z-score > 1.5 (high quality wines)?
high_quality_wines = len(winequality_df[winequality_df['quality_z_score'] >= 1.5])
print("1. The amount of high quality wine is", high_quality_wines)
print()

#2. How many wines have a Z-score < -1.5 (low quality wines)?
low_quality_wines = len(winequality_df[winequality_df['quality_z_score'] <= -1.5])
print("2. The amount of low quality wine is", low_quality_wines)
print()

#3. What is the percentage of the entire data?
high_quality_pct =  high_quality_wines / len(winequality_df) * 100
low_quality_pct = low_quality_wines / len(winequality_df) * 100
print(f"""3. The high quality wine percentage is {high_quality_pct:.2f}%
while the low quality percentage is {low_quality_pct:.2f}% """)
```

##Part III

```python
#take a subset of data with High Quality Wines
print("Subset data of High Quality Wines")
high_quality_wines = winequality_df[winequality_df['quality_z_score'] >= 1.5]
print(high_quality_wines.head())
```

```python
#take a subset of data with Low Quality Wines
print("Subset data of Low Quality Wines")
low_quality_wines = winequality_df[winequality_df['quality_z_score'] <= -1.5]
print(low_quality_wines.head())
```

##Part IV

```python
#Compare the descriptive statistics each subset with alcohol, volatile acidity, residual sugar, sulphates, citric acid
high_quality_wines = winequality_df[winequality_df['quality_z_score'] >= 1.5]
low_quality_wines = winequality_df[winequality_df['quality_z_score'] <= -1.5]
features = ["alcohol", "volatile acidity", "residual sugar", "sulphates", "citric acid"]

print("High Quality Wines Descriptive Statistics:")
print()
print(high_quality_wines[features].describe())
print()
print()
print("Low Quality Wines Descriptive Statistics:")
print()
print(low_quality_wines[features].describe())
```

##Part V

```python
#Analytical Thinking Answers
#1.Is the alcohol content generally higher in high-quality wines?
print("1.")
print(f"Mean Alcohol in High Quality Wines = {high_quality_wines['alcohol'].mean():.2f}")
print(f"Mean Alcohol in Low Quality Wines = {low_quality_wines['alcohol'].mean():.2f}")

if high_quality_wines['alcohol'].mean() > low_quality_wines['alcohol'].mean():
  print("The alcohol content is higher in High Quality Wines")
else:
  print("The alcohol content is higher in Low Quality Wines")
```

```python
#2.Is volatile acidity lower in high-quality wine?
print("2.")
print(f"Mean Volatile Acidity in High Quality Wines = {high_quality_wines['volatile acidity'].mean():.2f}")
print(f"Mean Volatile Acidity in Low Quality Wines = {low_quality_wines['volatile acidity'].mean():.2f}")

if high_quality_wines['volatile acidity'].mean() < low_quality_wines['volatile acidity'].mean():
  print("Volatile acidity is lower in High Quality Wines")
else:
  print("Volatile acidity is lower in Low Quality Wines")
```

```python
#3. Which feature most consistently distinguishes quality? What makes wine high or low quality?
features = ["alcohol", "volatile acidity", "residual sugar", "sulphates", "citric acid"]

data = pd.DataFrame({
    "High Quality Mean" : [high_quality_wines[ft].mean() for ft in features],
    "Low Quality Mean" : [low_quality_wines[ft].mean() for ft in features],
    "High Quality Std" : [high_quality_wines[ft].std() for ft in features],
    "Low Quality Std" : [low_quality_wines[ft].std() for ft in features]
}, index=features)

data["Difference"] = data["High Quality Mean"] - data["Low Quality Mean"]

data["Avg Std"] = (data["High Quality Std"] + data["Low Quality Std"]) / 2

print("3.")
print(data)

most_distinctive_index = data["Difference"].abs().idxmax()
most_distinctive = data.loc[most_distinctive_index]

print()
print(f"The most consistent feature that affects wine quality is {most_distinctive_index} with diff = {most_distinctive['Difference']:.2f}")
```

```python
#4. Which feature is the least consistent in deciding wine quality?
print("4.")
print(data)

least_distinctive_index = data["Difference"].abs().idxmin()
least_distinctive = data.loc[least_distinctive_index]

print()
print(f"The least consistent feature in deciding wine quality is {least_distinctive_index} with diff = {least_distinctive['Difference']:.2f}")
```

```python
print("Analytical Thinking Summary/Insight")
print("""Based on the analysis above, higher alcohol content is related to higher wine quality,
while lower volatile acidity is also related to better wine quality.
Among the features, alcohol content shows the most consistent difference,
meaning that alcohol content has a stronger impact on wine quality.
Meanwhile, residual sugar shows very little difference, indicating that it has a weak effect on wine quality
""")
```
